```python
class SRentalShop:
    def __init__(self,scooter_total): # the rental shop itself and how do we define itself
        self.scooter_total = scooter_total # initalizaion of available scooters
        self.rent_rates = {'hour': 7, 'day': 25, 'week': 70} # initalizaitons of rent rates
        self.rented_now = {} #defining the list of scooters rented out
    
    def rented_out_scooters(self,customer_name,scooter_need, rental_time,time_length):
    #Handle Scooter Rentals
        # We are using if statements so it will only be run once we don't want to be looping forever and     
        if scooter_need > self.scooter_total:
            return f"Error: Only {self.scooter_total} scooters available for rent."
        if rental_time not in self.rent_rates:
            return "Error: Invalid rental time. Choose 'hour', 'day', or 'week'."       
    #Take Customer Details
        transactions = {
            'scooter_need': scooter_need,
            'rental_time': rental_time,
            'time_length': time_length,
            'amount_charged': 0 
        }
        self.rented_now[customer_name] = transactions
        self.scooter_total -= scooter_need
        return f"{scooter_need} scooters rented to {customer_name} for {time_length} {rental_time}."


        #Scooter Return Process
    
    def returned_scooters(self, customer_name):

        if customer_name not in self.rented_now:
            return "Error: No transaction found for this customer."
        
        transactions = self.rented_now[customer_name]
        scooter_need = transactions['scooter_need']
        rental_time = transactions['rental_time']
        time_length = transactions['time_length']
        
        self.scooter_total += scooter_need
        start_cost = self.rent_rates[rental_time] * time_length * scooter_need
        
        if 3 <= scooter_need <= 5:
            discount = 0.3 * start_cost
        else:
            discount = 0
        
        amount_charged = start_cost - discount
        del self.rented_now[customer_name]  # Remove the transaction record
        return f"Total amount due for {customer_name}: ${amount_charged:.2f}" # .2f allows it to be in decimal and cents format
# Unit Test
if __name__ == "__main__":
    shop = SRentalShop(scooter_total=10)
    
    # Test Case 1: Successful rental
    print(shop.rented_out_scooters("Ronald Rogers", 3, 'day', 2))  # Expected: 3 scooters rented to Ronald Rogers for 2 day(s).
    print(shop.returned_scooters("Ronald Rogers"))  # Expected: Total amount due for Ronald Rogers: $105.00 (since 2 days * $25 * 3 scooters * 0.7 discount)

    # # Test Case 2: Rental with invalid time
    print(shop.rented_out_scooters("Quincy Quinoa", 1, 'month', 1))  # Expected: Error: Invalid rental period. Choose 'hour', 'day', or 'week'.

    # # Test Case 3: Rental more than avaliable
    print(shop.rented_out_scooters("Alex Armadillo", 11, 'hour', 3))  # Expected: Error: Only 10 scooters available for rent.

    # # Test Case 4: Successful rental and return no discount
    print(shop.rented_out_scooters("Macy Machintosh", 2, 'week', 1))  # Expected: 2 scooters rented to Macy Machintosh for 1 week(s).
    print(shop.returned_scooters("Macy Machintosh"))  # Expected: Total amount due for Macy Machintosh: $140.00

    # # Test Case 5: No transaction found
    print(shop.returned_scooters("Non-existent Customer"))  # Expected: Error: No transaction found for this customer.
    
    # # Test Case 6: Emergency stop
    print("Ending program.")
```
